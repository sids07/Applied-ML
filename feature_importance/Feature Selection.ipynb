### Feature Selection

```python
import pandas as pd

df = pd.read_csv('mobile-train.csv')
df.head()
```

#### Univariate Selection

```python
X = df.iloc[:,:-1]
y = df['price_range']
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.feature_selection import SelectKBest #also called information gain
from sklearn.feature_selection import chi2
```

```python
df.shape #20 features
```

```python
order_rank_feature = SelectKBest(score_func=chi2,k=10)
ordered_feature=order_rank_feature.fit(X,y)
df_scores=pd.DataFrame(ordered_feature.scores_,columns=['Score'])
df_columns = pd.DataFrame(df.columns)
```

```python
scores = pd.concat([df_columns,df_scores],axis=1)
scores.columns = ['Features','Scores']
scores
```

```python
scores.nlargest(10,'Scores')
```

#### Feature Importance

Gices us a score for each features higher the score of feature more relevant

```python
from sklearn.ensemble import ExtraTreesClassifier
import matplotlib.pyplot as plt
model = ExtraTreesClassifier()
model.fit(X,y)
```

```python
print(model.feature_importances_)
```

```python
ranked_features=pd.Series(model.feature_importances_,index= X.columns)
ranked_features.nlargest(10).plot(kind='barh')
plt.show()
```

#### Correlation

```python
import seaborn as sns
corr= df.iloc[:,:-1].corr()
corr
```

```python
top_feature = corr.index
plt.figure(figsize=(20,20))
sns.heatmap(df[top_feature].corr(),annot=True)
```

```python
threshold = 0.5

def correlation(dataset,threshold):
    col_corr = set()
    corr_mat = dataset.corr()
    
    for i in range(len(corr_mat.columns)):
        for j in range(i):
            if abs(corr_mat.iloc[i,j])>threshold:
                colname = corr_mat.columns[i]
                col_corr.add(colname)
    return col_corr
    
```

```python
correlation(df.iloc[:,:-1],threshold)
```

Here we can use any of the correlated feature.

Why?

Because they are independent feature which will effect dependent features. They cant be correlated otherwise they will be almost same feature i.e. duplicate features

### Information Gain

```python
from sklearn.feature_selection import mutual_info_classif
```

```python
mutual_info = mutual_info_classif(X,y)
```

```python
mutual_data=pd.Series(mutual_info,index=X.columns)
mutual_data.sort_values(ascending=False)
```

##### postive number which measures are dependices. It is equalto zero if and only if two random variables are independent, and higher values mean higher dependency.
